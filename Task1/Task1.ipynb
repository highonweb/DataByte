```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python

data = pd.read_csv("./company_sales_data - company_sales_data.csv", index_col="month_number")
```

```python
data.head()
```

```python

plt.figure(figsize=(20,10))
plt.title("Trend of total profits")
ax = sns.lineplot(data=data[['total_profit']],dashes=False)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
```

```python
plt.figure(figsize=(20,10))
plt.title("Trend of individual product profits")
ax = sns.lineplot(data=data.iloc[:,:-2],dashes=False)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
```

```python
Total = data[list(data.columns)[:-2]].sum()
```

```python
pie, ax = plt.subplots(figsize=[10,6])
plt.pie(x=Total, autopct="%.1f%%",labels=list(data.columns)[:-2], explode=[0.05]*len(list(data.columns)[:-2]), pctdistance=0.5)
plt.title("Total profits by product", fontsize=14);

```
