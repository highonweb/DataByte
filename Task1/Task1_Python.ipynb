```python
import csv

#getting the country codes
code1 = input("first country code ")
code2 = input("second country code ")

#opening csv file
with open('./data_csv.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    csv_reader = list(sorted(csv_reader, key=lambda row: row['Code']))
    
    # flag for saying whether the loop should print the country names
    flag=False
    
    for row in csv_reader:
        
        #toggling the flag if any of the code matches country code
        if(row["Code"] == code1 or row["Code"] == code2 ):
            flag = not flag
            
        #Printing the country name and excluding the both given countries
        if(flag and row["Code"] != code1 and row["Code"] != code2):
            print(row["Name"])
        
```
